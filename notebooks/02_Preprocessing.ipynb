```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from pathlib import Path

from sklearn.model_selection import train_test_split
```

```python
PROJECT_ROOT = Path("..").resolve()

DATA_PATH = PROJECT_ROOT / "data" / "raw" / "ai4i2020.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(
    columns=[
        "UDI",
        "Product ID"
    ],
    inplace=True
)
```

```python
df.columns = (
    df.columns
      .str.replace("[","",regex=False)
      .str.replace("]","",regex=False)
      .str.replace(" ","_")
      .str.replace("/","_")
)
```

```python
df.head()
```

```python
df = pd.get_dummies(
    df,
    columns=["Type"],
    drop_first=False,
    dtype=int
)
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
X = df.drop(
    columns=["Machine_failure"]
)

y = df["Machine_failure"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

PROCESSED_DIR.mkdir(exist_ok=True)
```

```python
train = X_train.copy()

train["Machine_failure"] = y_train.values

test = X_test.copy()

test["Machine_failure"] = y_test.values
```

```python
train.to_csv(
    PROCESSED_DIR / "train.csv",
    index=False
)

test.to_csv(
    PROCESSED_DIR / "test.csv",
    index=False
)
```
